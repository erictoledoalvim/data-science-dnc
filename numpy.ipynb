```python
# Instalando biblioteca de leitura de imagens de satélite 'rasterio'
%pip install rasterio
```

```python
# Importando biblioteca de leitura de imagens de satélite
import rasterio
from rasterio.plot import show

# Abrindo visualizador de imgem
dataset = rasterio.open('/Users/Eric/Workspace/data-science-dnc/crop_rapideye.tif')
show(dataset)

# Salvando as informações da imagem como uma matriz
dataset_array = dataset.read(1)

# Exibindo o tamanho da matriz
print(dataset_array.shape)

# Exibindo as dimensões da matriz
print(dataset_array.ndim)

# Exibindo o tipo de dado que esta armazenado
print(dataset_array.dtype)

# Exibindo o tamanho do elemento em bytes
print(dataset_array.itemsize)

# Numero total de elementos da array
print(dataset_array.size)
```

# Aula 6

```python
# Estatística descritiva de uma array numpy
# Calculando a média
print(dataset_array.mean())

# Calculando o desvio padrão
print(dataset_array.std())

# Calculando o menor valor
print(dataset_array.min())

# Calculando o maior valor
print(dataset_array.max())

# Soma total dos valores brutos
print(dataset_array.sum())

# Identificando os valores únicos
unique_values = np.unique(dataset_array)
print(unique_values)
```
