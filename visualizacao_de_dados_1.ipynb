# Aula 1

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
```

```python
credit = pd.read_excel('clientes_renamed.xlsx')
```

```python
# Agrupando dados
credit_group = credit.groupby(['EDUCAÇÃO']).mean().reset_index()
credit_group
```

```python
# MATPLOTLIB: média do valor de crédito concedido segundo o nivel de escolaridade
plt.figure(figsize = (8,8))
plt.xlabel('Nível de escolaridade')
plt.ylabel('Média do valor de crédito concedido')
plt.title('Valor de crédito concedido segundo o nível de escolaridade')
plt.bar(credit_group['EDUCAÇÃO'],credit_group['VALOR_CREDITO'], color = 'red');
# para tirar o "<BarContainer object of 7 artists>" uso o ; no fim do plt.bar
```

# Aula 2

```python
# SEABORN: extrato bancário dos clientes segundo o nível de escolaridade para os últimos 4 meses
fig, axis = plt.subplots(2,2,figsize = (15,15))
sns.barplot(data = credit_group, x = 'EDUCAÇÃO', y = 'EXTRATO_JUN_2005', ax = axis[0,0])
sns.barplot(data = credit_group, x = 'EDUCAÇÃO', y = 'EXTRATO_JUL_2005', ax = axis[0,1])
sns.barplot(data = credit_group, x = 'EDUCAÇÃO', y = 'EXTRATO_AGO_2005', ax = axis[1,0])
sns.barplot(data = credit_group, x = 'EDUCAÇÃO', y = 'EXTRATO_SET_2005', ax = axis[1,1])
```

# Aula 3

```python
# PLOTLY:

fig = go.Figure()

fig.add_trace(go.Bar(x = credit_group['EDUCAÇÃO'], y = credit_group['VALOR_PAG_ABR_2005'], marker_color = 'Red', name = 'Valor pago 04/2005'))
fig.add_trace(go.Bar(x = credit_group['EDUCAÇÃO'], y = credit_group['VALOR_PAG_MAI_2005'], marker_color = 'Blue', name = 'Valor pago 05/2005'))
fig.add_trace(go.Bar(x = credit_group['EDUCAÇÃO'], y = credit_group['VALOR_PAG_JUN_2005'], marker_color = 'Yellow', name = 'Valor pago 06/2005'))
fig.add_trace(go.Bar(x = credit_group['EDUCAÇÃO'], y = credit_group['VALOR_PAG_JUL_2005'], marker_color = 'Green', name = 'Valor pago 07/2005'))
fig.add_trace(go.Bar(x = credit_group['EDUCAÇÃO'], y = credit_group['VALOR_PAG_AGO_2005'], marker_color = 'Purple', name = 'Valor pago 08/2005'))
fig.add_trace(go.Bar(x = credit_group['EDUCAÇÃO'], y = credit_group['VALOR_PAG_SET_2005'], marker_color = 'Orange', name = 'Valor pago 09/2005'))

fig.update_layout(yaxis = dict(title = 'Valor pagamento em 2005', titlefont_size = 14, tickfont_size = 10),
                  xaxis = dict(title = 'Nível de escolaridade', titlefont_size = 14, tickfont_size = 10))

# Salvando o gráfico interativo
# fig.write_html('valor_pago_escolaridade.html')
```

# Aula 4

```python
# Agrupando dados conforme idade
credit_group = credit.groupby(['IDADE']).mean().reset_index()
credit_group
```

```python
# MATPLOTLIB
plt.figure(figsize = (10,8))

plt.title('Valor de crédito concedido aos clientes segundo a idade')
plt.xlabel('Idade')
plt.ylabel('Média do valor de crédito concedido')
plt.grid(visible = True, color = 'gray', linestyle = '--', linewidth = 0.5)

plt.plot(credit_group['IDADE'],credit_group['VALOR_CREDITO']);
```

# Aula 5

```python
# GRAFICO COM A SEABORN

fig, axis = plt.subplots(2,2, figsize = (20,10))

sns.lineplot(data = credit, x = 'IDADE', y = 'EXTRATO_JUN_2005', hue = 'SEXO', style = 'SEXO', palette = 'Set2', ax = axis[0,0]).set(title = 'EXTRATO EM JUNHO DE 2005', xlabel = None)
sns.lineplot(data = credit, x = 'IDADE', y = 'EXTRATO_JUL_2005', hue = 'SEXO', style = 'SEXO', palette = 'Set2', ax = axis[0,1]).set(title = 'EXTRATO EM JULHO DE 2005', xlabel = None)
sns.lineplot(data = credit, x = 'IDADE', y = 'EXTRATO_AGO_2005', hue = 'SEXO', style = 'SEXO', palette = 'Set2', ax = axis[1,0]).set(title = 'EXTRATO EM AGOSTO DE 2005')
sns.lineplot(data = credit, x = 'IDADE', y = 'EXTRATO_SET_2005', hue = 'SEXO', style = 'SEXO', palette = 'Set2', ax = axis[1,1]).set(title = 'EXTRATO EM SETEMBRO DE 2005')
```

```python
# GRÁFICO COM A PLOTLY: valor pago ao longo dos meses de acordo com a idade
fig = go.Figure()
from plotly.subplots import make_subplots
fig = make_subplots(rows = 3, cols = 2, shared_yaxes = True, subplot_titles = ('Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro'))
fig.add_trace(go.Scatter(   x = credit_group['IDADE'],y = credit_group['VALOR_PAG_ABR_2005'],line = dict(color = 'red', width = 1.5), name = 'Valor pago 04/2005'), row = 1, col = 1)
fig.add_trace(go.Scatter(   x = credit_group['IDADE'],y = credit_group['VALOR_PAG_MAI_2005'],line = dict(color = 'blue', width = 1.5), name = 'Valor pago 05/2005'), row = 1, col = 2)
fig.add_trace(go.Scatter(   x = credit_group['IDADE'],y = credit_group['VALOR_PAG_JUN_2005'],line = dict(color = 'green', width = 1.5), name = 'Valor pago 06/2005'), row = 2, col = 1)
fig.add_trace(go.Scatter(   x = credit_group['IDADE'],y = credit_group['VALOR_PAG_JUL_2005'],line = dict(color = 'orange', width = 1.5), name = 'Valor pago 07/2005'), row = 2, col = 2)
fig.add_trace(go.Scatter(   x = credit_group['IDADE'],y = credit_group['VALOR_PAG_AGO_2005'],line = dict(color = '#FF0092', width = 1.5), name = 'Valor pago 06/2005'), row = 3, col = 1)
fig.add_trace(go.Scatter(   x = credit_group['IDADE'],y = credit_group['VALOR_PAG_SET_2005'],line = dict(color = 'purple', width = 1.5), name = 'Valor pago 07/2005'), row = 3, col = 2)

fig.update_layout(title_text = 'Valor pago em 2005 ao longo dos meses de acoro com a idade dos clientes')
fig.update_yaxes(title_text = 'Valor pago em dólar NT', row = 1, col = 1, titlefont_size = 8)
fig.update_yaxes(title_text = 'Valor pago em dólar NT', row = 1, col = 2, titlefont_size = 8)
fig.update_yaxes(title_text = 'Valor pago em dólar NT', row = 2, col = 1, titlefont_size = 8)
fig.update_yaxes(title_text = 'Valor pago em dólar NT', row = 2, col = 2, titlefont_size = 8)
fig.update_yaxes(title_text = 'Valor pago em dólar NT', row = 3, col = 1, titlefont_size = 8)
fig.update_yaxes(title_text = 'Valor pago em dólar NT', row = 3, col = 2, titlefont_size = 8)

fig.update_xaxes(title_text = 'idade', row = 3, col = 1, titlefont_size = 8)
fig.update_xaxes(title_text = 'idade', row = 3, col = 2, titlefont_size = 8)


fig.write_html('Valor pago idade.html')
```

# Aula 6

```python
# Agrupando os dados
credit_group = credit.groupby(['SEXO']).mean().reset_index()
```

```python
# Gráfico com a MATPLOTLIB
# 1 - sexo masculino
# 2 - sexo feminino

fig, ax = plt.subplots(figsize = (8,8))

explode = (0.05,0)

ax.pie(credit_group['VALOR_CREDITO'], labels = credit_group['SEXO'], autopct = '%.2f%%', shadow = True, startangle = 60, colors = ['red','seagreen'], explode = explode);

ax.set_title('Porcentagem do valor de credito concedido segundo o sexo', fontsize = 14)
```

# Aula 7

```python
# Gráfico de Pizza com a Plotly
from plotly.subplots import make_subplots

fig = make_subplots(rows = 1, cols = 2, specs = [[{'type':'domain'},{'type':'domain'}]])

fig.add_trace(go.Pie(labels = credit_group['SEXO'], values = credit_group['EXTRATO_ABR_2005'], name = 'Abril'), 1, 1)
fig.add_trace(go.Pie(labels = credit_group['SEXO'], values = credit_group['EXTRATO_SET_2005'], name = 'Abril'), 1, 2)

fig.update_traces(hole = 0.3)
fig.update_layout(title_text = 'Extrato de pagamento (Abril e Setembro) de acordo com o sexo')

fig.write_html('extrato_sexo.html')
```
