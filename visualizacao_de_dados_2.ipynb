# Aula 1

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import numpy as np
```

```python
# Criando uma matriz com a funo linspace da numpy

t = np.linspace(0.0,2.0,100)

# Criar outras 2 matrizes com diferentes funes trigonomtricas

f1 = np.sin(np.pi*t)
f2 = np.sin(2 * np.pi * t)

plt.figure(figsize = (10,8))
plt.plot(t, f1, linewidth = 5, label = 'seno de pi * t')
plt.plot(t,f2, linewidth = 5, label = 'seno de 2 * pi * t')
plt.legend()
plt.title('SENOIDES')
plt.grid(b = True, color = 'gray', linestyle = '--', linewidth = 0.5)
```

```python
# SEABORN: DIAGRAMA DE CORRELAO

credit = pd.read_excel('clientes_renamed.xlsx')
```

```python
credit
```

```python
credit.drop(['Unnamed: 0', 'ID'], axis = 1, inplace = True)
```

```python
plt.figure(figsize = (20,20))
sns.heatmap(credit.corr(), annot = True, cmap = 'coolwarm')
```

# Aula 3

```python
# PLOTLY: VIOLIN (Histograma + boxplot)

fig = px.violin(credit, x = 'ESTADO_CIVIL', y = 'VALOR_CREDITO', color = 'SEXO', box = True,title = 'Valor de crdito concedido segundo o estado civil dos clientes')
fig.show()
```

# Aula 4

```python
# PLOTLY: Scatter

fig = px.scatter(credit, x = 'PAG_SET_2005', y = 'VALOR_PAG_SET_2005', color = 'ESTADO_CIVIL', size = 'VALOR_CREDITO', title = 'Sintese da situao dos clientes para o mes de setembro de 2005 segundo o estado civil')
fig.show()

fig.write_html('sintese_clientes_set_2005.html')
```
